```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import confusion_matrix
```

```python
train_df = pd.DataFrame.from_csv('train.csv')
valid_df = pd.DataFrame.from_csv('valid.csv')
```

```python
train_df
```

```python
valid_df
```

```python
(train_df['Survived'] == True).value_counts(True)
```

```python
(valid_df['Survived'] == True).value_counts(True)
```

```python
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(train_df[['Pclass', 'SibSp', 'Parch']], train_df['Survived'])
```

```python
predictions = classifier.predict(valid_df[['Pclass', 'SibSp', 'Parch']])
df = predictions == valid_df['Survived']
df.value_counts(True)
```

```python
confusion_matrix(valid_df['Survived'], predictions) / len(valid_df)
```
